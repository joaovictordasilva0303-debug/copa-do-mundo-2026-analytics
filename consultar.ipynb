```python
conn = sqlite3.connect("banco.db")

query = """
    SELECT selecao, SUM(xg) AS xg    
    FROM partidas
    GROUP BY selecao
    HAVING SUM(xg) > (
        SELECT AVG(total)
        FROM(
            SELECT selecao, SUM(xg) AS total
            FROM partidas
            GROUP BY selecao
        )t
    ) ORDER BY xg DESC
;
"""

df_copa = pd.read_sql_query(query, conn)
display(df_copa)
```
